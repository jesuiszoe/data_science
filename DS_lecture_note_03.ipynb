## 정규 표현식 (Regular Expression)

- 정규 표현식은 텍스트 문자열을 패턴으로 파악하여 식별하는데 쓰인다. 즉, 문자열이 주어진 규칙에 일치하는지, 일치하지 않는지 판단할 수 있다.

Example) 신문기사 본문을 수집할 경우 본문에 기자의 이메일 주소가 들어가 있는 경우가 많다. 실제 텍스트 자료를 분석 시 이러한 내용은 불필요하므로 제거해야 하는데 이때 정규 표현식을 활용해 이메일 주소 패턴에 일치하는 문자열을 찾아 삭제할 수 있다.

## 전처리 (Text preprocessing)

- 텍스트 마이닝에 적용되는 텍스트 데이터 집합을 '말뭉치(corpus)'라고 부른다. 이는 대용량의 정형화된 텍스트 집단으로 정의되는데, '정형화된'이란 뜻은 원(raw) 텍스트 데이터가 사전 처리 작업이 되었음을 의미한다.
<br><br>
- 텍스트 데이터의 전처리 과정은 특정한 법칙이나 규정은 없지만, 일반적으로 몇 가지 공통적인 과정을 거친다.

### (1) 대소문자 통일

### (2) 숫자, 문장부호, 특수문자 제거

- 원 데이터에는 텍스트 뿐만 아니라 숫자, 문장부호, 특수문자 등이 모두 포함되어 있다. 이 것들은 대부분 분석에 포함되지 않기 때문에 제거하는 과정이 필요하다. (단, 경우에 따라 필요한 숫자, 문장부호, 특수문자 등이 존재하기 때문에 연구자의 판단에 따라 제거 대상을 선별하는 경우도 있다.)

```python
r_number = re.compile('[0-9]+') # 숫자 제거
r_symbol = re.compile('\W+') # 특수문자 제거
r_underbar = re.compile('_') # 추가적으로 제거해야 하는 기호가 있을 경우
```

```python
string = '문재인정부의 전체 887개 공약 중 완료된 것은 155건으로 17.47%의 이행률을 기록 중인 것으로 나타났다.'
print(string)
```

```python
string = r_number.sub(' ', string)
print(string)
```

```python
string = r_symbol.sub(' ', string)
print(string)
```

### (3) 불용어 (stopword) 제거

- 텍스트 데이터에 자주 사용되지만 특별한 의미를 부여하기 힘든 단어를 '불용어(stopword)'라고 부른다. (예를들면 영어의 'the', 'a', 'an'과 같은 관사)
<br><br>
- 파이썬에서 가장 많이 쓰는 불용어 리스트 패키지는 'NLTK'인데 아쉽게도 한국어의 경우 아직 지원하지 않는다. 따라서 한국어의 경우는 불용어 리스트를 직접 작성해 사용하게 된다.

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
# 한국어 불용어 제거 방법
words_korean = ['추석', '연휴', '민족', '대이동', '시작', '늘어', '교통량', '교통사고', 
                '특히', '자동차', '고장', '상당수', '차지', '나타', '것', '기자']
print(words_korean)
```

## 품사 분석 (Part Of Speech, POS)

- 문장을 구성하는 개별 단어는 문장 안에서 고유한 문법적 기능을 수행한다.
<br><br>
- 동일한 단어라도 서로 다른 품사로 쓰이는 경우가 있다. 보통의 경우 같은 단어를 품사와 관계없이 동일한 형태의 단어로 보고 분석하는 경우가 있지만, 경우에 따라 특정 품사만을 수집해 분석하고 싶은 경우도 있다. (예를들어 문서의 감정 정도를 파악하고 싶은 경우 형용사만 추출하면 더욱 효율적으로 파악할 수 있다.) 기본적인 텍스트 마이닝은 문서에서 명사만을 추출해서 분석하는 경우가 대부분이다.
<br><br>
- 한국어 품사 분석을 위해 KoNLPy 패키지를 사용한다. 이 패키지 안에서 Kkma, Komoran, Hannaum, Twitter, Mecab 클래스를 이용해 품사 분석을 할 수 있다. 실행 시간과 정확도에 대한 판단 기준을 통해 클래스를 선택하면 된다.
<br><br>
- 영어의 경우 NLTK 패키지를 사용하면 된다.

```python
from IPython.display import Image
```

```python
# 한국어 품사분석
from konlpy.tag import Hannanum
from konlpy.tag import Kkma
from konlpy.tag import Twitter
from konlpy.tag import Okt # KoNLPy v0.4.5. 이후에는 "Twitter" 클래스 명이 "Okt"로 변경

string_kor = "친척들이 모인 이번 추석 차례상에서는 단연 '집값'이 화제에 올랐다."
```

- morphs : 문장을 형태소 단위로 끊어주는 매서드
- nouns : 명사를 추출해주는 매서드
- pos : 각 형태소의 품사까지 알려주는 매서드

```python
# Hannanum
hannanum = Hannanum()
morphs_hannanum = hannanum.morphs(string_kor)
nouns_hannanum = hannanum.nouns(string_kor)
pos_hannanum = hannanum.pos(string_kor)
print(morphs_hannanum)
print(nouns_hannanum)
print(pos_hannanum)
```

```python
# Kkma
kkma = Kkma()
morphs_kkma = kkma.morphs(string_kor)
nouns_kkma = kkma.nouns(string_kor)
pos_kkma = kkma.pos(string_kor)
print(morphs_kkma)
print(nouns_kkma)
print(pos_kkma)
```

```python
# Twitter, KoNLPy v0.4.5. 이후에는 "Twitter" 클래스 명이 "Okt"로 변경
twitter = Twitter()
morphs_twitter = twitter.morphs(string_kor)
nouns_twitter = twitter.nouns(string_kor)
pos_twitter = twitter.pos(string_kor)
print(morphs_twitter)
print(nouns_twitter)
print(pos_twitter)
```

```python
# Okt
okt = Okt()
morphs_okt = okt.morphs(string_kor)
nouns_okt = okt.nouns(string_kor)
pos_okt = okt.pos(string_kor)
print(morphs_okt)
print(nouns_okt)
print(pos_okt)
```

## 단어 빈도분석 (Word Frequency)

- 텍스트 데이터에 대한 빈도분석은 가장 기본적이지만 쉽고 보편적으로 활용되는 방법이며, 본격적인 분석 이전에 데이터에 대한 이해와 흐름을 살펴보기 위한 기초 분석에 해당한다. (기존 수치 데이터 분석 이전 평균, 표준편차 등을 구하는 것과 유사하다.)

```python
# 한글 단어 빈도분석
import pandas as pd
from konlpy.tag import Hannanum
from konlpy.tag import Kkma
hannanum = Hannanum()
kkma = Kkma()
```

```python
f_kor = open('/Users/zoe/Downloads/moon.txt', 'r', encoding='cp949')
lines_kor = f_kor.readlines()
f_kor.close()
print(lines_kor)
```

```python
temp_hannanum = []
temp_kkma = []

for i in range(len(lines_kor)):
    temp_hannanum.append(hannanum.nouns(lines_kor[i]))
    temp_kkma.append(kkma.nouns(lines_kor[i]))
```

```python
print(temp_hannanum)
```

```python
print(temp_kkma)
```

```python
# 각 line들을 합쳐주는 함수 작성
def flatten(l):
    flatList = []
    for elem in l:
        if type(elem) == list:
            for e in elem:
                flatList.append(e)
        else:
            flatList.append(elem)
    return flatList
```

```python
word_list_hannanum = flatten(temp_hannanum)
print(word_list_hannanum)
```

```python
word_list_kkma = flatten(temp_kkma)
print(word_list_kkma)
```

```python
word_list_hannanum = pd.Series([x for x in word_list_hannanum if len(x)>1])
pd.DataFrame(word_list_hannanum)
```

```python
word_list_kkma = pd.Series([x for x in word_list_kkma if len(x)>1])
pd.DataFrame(word_list_kkma)
```

```python
word_list_hannanum_20 = word_list_hannanum.value_counts().head(20)
pd.DataFrame(word_list_hannanum_20)
```

```python
word_list_kkma_20 = word_list_kkma.value_counts().head(20)
pd.DataFrame(word_list_kkma_20)
```
